```
import pymysql

# Question 1 Insert
pydb = pymysql.connect("localhost", "root", "Update12!", "pydb")
dbcursor = pydb.cursor()

def insert_student():
    stuid = input("Please enter an id: ")
    stuname = input("Please enter a name: ")
    major = input("Please enter a major: ")
    stuyear = int(input("Please enter starting year: "))
    tuition = input("Please enter tuition: ")
    gradyear = int(input("Please enter year to graduate: "))

    sql = "INSERT INTO tstudent(stuid,stuname,major,stuyear,tuition,gradyear) VALUES(%s,%s,%s,%s,%s,%s)"
    if not stuid.isnumeric():
        print("The id is incorrect")
        exit()
    else:
        values=(stuid,stuname,major,stuyear,tuition,gradyear)
        command=dbcursor.execute(sql,values)
        pydb.commit()
        print(dbcursor.rowcount, "record(s) affected")
        dbcursor.close()
        pydb.close()

insert_student()

#Question 2 Delete
pydb = pymysql.connect("localhost", "root", "Update12!", "pydb")
dbcursor = pydb.cursor()

def delete_student_by_id():
    sql ="DELETE FROM tstudent WHERE stuid=%s"
    value=int(input("Please input an id you want to delete: "))
    dbcursor.execute(sql,value)
    pydb.commit()
    dbcursor.close()
    pydb.close()

delete_student_by_id()

#Question 3 Update

pydb = pymysql.connect("localhost", "root", "Update12!", "pydb")
dbcursor = pydb.cursor()

def update_student():
    stuid = input("Please enter id of student: ")
    stuname = input("Please enter name of student: ")
    stuname_new = input("Please enter new name of student: ")
    sql ="UPDATE tstudent SET stuname=%s WHERE stuname=%s AND stuid=%s"
    if not stuid.isnumeric():
        print("The id is incorrect")
        exit()
    else:
        values=(stuname_new,stuname,stuid)
        command=dbcursor.execute(sql,values)
        pydb.commit()
        print(dbcursor.rowcount, "record(s) affected")
        dbcursor.close()
        pydb.close()

update_student()



#Question 4

pydb = pymysql.connect("localhost", "root", "Update12!", "pydb")
dbcursor = pydb.cursor()

def count_students_in_major():
    sql = "select major, count(*) from tstudent where major = %s group by major having count(*) > 1"
    value = input("Please enter the major (output count of students in major): ")
    command=dbcursor.execute(sql,value)
    results=dbcursor.fetchall()
    print(results)

    pydb.commit()
    dbcursor.close()
    pydb.close()

count_students_in_major()



#Question 5

pydb = pymysql.connect("localhost", "root", "Update12!", "pydb")
dbcursor = pydb.cursor()

def random_id_name():
    sql = "select stuid, stuname from tstudent order by rand() limit 1"
    command=dbcursor.execute(sql)
    results=dbcursor.fetchall() #query limits by one so no need to use fetchone()
    print(results)

    pydb.commit()
    dbcursor.close()
    pydb.close()

random_id_name()
```
